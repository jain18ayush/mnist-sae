```python
import matplotlib.pyplot as plt
import seaborn as sns
import torch

for depth in range(1, 9):
    # Load the activation tensor
    activations = torch.load(f'{prefix}embeddings/mnist-colored_sae_MNIST_depth_{depth}.pth')
    print(f"Activations shape for depth {depth}: {activations.shape}")

    # Transpose to get neurons in the correct dimension
    transposed = activations.t()

    # Get the top 10 maximum values per neuron along dimension 1
    top_activations, _ = torch.topk(transposed, k=10, dim=1)
    print(f"Top activations shape for depth {depth}: {top_activations.shape}")

    # Flatten the top activations for histogram plotting
    top_activations_flat = top_activations.flatten().numpy()

    # Plotting
    plt.figure(figsize=(10, 6))
    sns.histplot(top_activations_flat, kde=True, bins=30)
    plt.xlabel('Top Activation Values')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Top 10 Activation Values per Neuron for Depth {depth}')
    # plt.savefig(f'plots/top_10_activations_per_neuron_MNIST_fc1_depth_{depth}.png')
    plt.show()
    plt.close()
```

# Analysis

Max_Value || Neuron_Index || Predicted_Label

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is the dataframe with the following columns: 'Max_Value', 'Neuron_Index', 'Predicted_Label'
# e.g. df = pd.DataFrame({'Max_Value': [...], 'Neuron_Index': [...], 'Predicted_Label': [...]})

# 1. Frequency of Neuron Activations per Label
def neuron_activation_frequency(df):
    freq_table = df.groupby(['Predicted_Label', 'Neuron_Index']).size().unstack(fill_value=0)
    return freq_table

# 2. Average Max Activation per Neuron per Label
def avg_max_activation_per_neuron(df):
    avg_table = df.groupby(['Predicted_Label', 'Neuron_Index'])['Max_Value'].mean().unstack(fill_value=0)
    return avg_table

# 3. Distribution of Max Activations
def plot_activation_distribution(df):
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Max_Value'], kde=True, bins=30)
    plt.title('Distribution of Max Activation Values')
    plt.xlabel('Max Activation Value')
    plt.ylabel('Frequency')
    plt.show()

# 4. Neuron-Label Correlation
def neuron_label_correlation(df):
    correlation = df[['Neuron_Index', 'Predicted_Label']].corr()
    return correlation

# Plotting Functions

# Heatmap of Neuron Activation Frequency
def plot_neuron_activation_heatmap(freq_table):
    plt.figure(figsize=(12, 8))
    sns.heatmap(freq_table, cmap="Blues", annot=False, cbar=True)
    plt.title('Neuron Activation Frequency by Label')
    plt.xlabel('Neuron Index')
    plt.ylabel('Predicted Label')
    plt.show()

# Heatmap of Average Max Activation per Neuron
def plot_avg_max_activation_heatmap(avg_table):
    plt.figure(figsize=(12, 8))
    sns.heatmap(avg_table, cmap="YlGnBu", annot=False, cbar=True)
    plt.title('Average Max Activation per Neuron by Label')
    plt.xlabel('Neuron Index')
    plt.ylabel('Predicted Label')
    plt.show()

# 5. Execute Analysis
def analyze_neurons(df):
    # Neuron activation frequency
    freq_table = neuron_activation_frequency(df)
    print("Neuron Activation Frequency Table:\n", freq_table)
    
    # Plot neuron activation heatmap
    plot_neuron_activation_heatmap(freq_table)
    
    # Average max activation per neuron
    avg_table = avg_max_activation_per_neuron(df)
    print("Average Max Activation per Neuron Table:\n", avg_table)
    
    # Plot average max activation heatmap
    plot_avg_max_activation_heatmap(avg_table)
    
    # Activation distribution
    plot_activation_distribution(df)
    
    # Neuron-label correlation
    correlation = neuron_label_correlation(df)
    print("Neuron-Label Correlation Matrix:\n", correlation)
```

```python
MNIST_df = pd.read_csv('docs/MNIST_results.csv')
analyze_neurons(MNIST_df)
```

```python
SAE_df = pd.read_csv('docs/SAE_results.csv')
analyze_neurons(SAE_df)
```
